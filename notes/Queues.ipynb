# Queues Using Linked Lists

```python
# Node class

class Node:
  def __init__(self, value):
    self.data = value
    self.next = None
```

```python
class QueueLL:
  # 1. Initialization
  def __init__(self):
    self.front = None
    self.rear = None


  # 2. Enqueue (Insertion)
  # Insertion always happens at rear
  def enqueue(self, value):
    new_node = Node(value)

    if self.rear == None:
      # Means queue empty h

      self.front = new_node
      self.rear = self.front

    else:
      # Means already items h queue m

      self.rear.next = new_node
      self.rear = new_node


  # 3. Dequeue (Deletion)
  # Deletion always happens at front
  def dequeue(self):

    if self.front == None:
      # Means queue is empty
      return "Empty Queue"

    else:
      self.front = self.front.next
      # Front k next vale ko front bnado


  # 4. Traverse
  def traverse(self):
    temp = self.front
    #Iterator

    while temp != None:
      print(temp.data, end=' ')
      temp = temp.next
```

# Queues Using 2 Stacks

```python
# Stacks
class StackLL:
  # 1. Initialization
  def __init__(self):
    self.top = None
    # Jese LL m head hota, vese stack m top hota


  # 2. Check if stack is empty
  def isEmpty(self):
    return self.top == None
    # Agar stack empty hua toh True return hojaega


  # 3. Push
  def push(self, value):
    new_item = Node(value)
    new_item.next = self.top
    self.top = new_item
    # Naya node banaya, Usko head/top ki taraf point kardia, Fir new node ko top banadia


  # 4. Pop
  def pop(self):
    # In case of empty stack
    if self.isEmpty():
      return "Stack empty"

    else:
      temp = self.top
      self.top = self.top.next
      return temp.data


  # 5. Traversal
  def traverse(self):
    # In case of empty stack
    if self.isEmpty():
      return "Stack empty"

    temp = self.top
    # Iterator

    while temp != None:
      print(temp.data)
      temp = temp.next


  # 6. Peek
  def peek(self):
    # In case of empty stack
    if self.isEmpty():
      return "Stack empty"
    else:
      return self.top.data
```

```python
# Queue using 2 stacks

# Use two stacks: stack1 and stack2.
## enqueue(x) → push x onto stack1. (O(1))
## dequeue() → if stack2 is non-empty, pop from stack2. Otherwise, move all elements from stack1 to stack2 (reverses order) and pop from stack2. This gives FIFO behaviour. (Amortized O(1))
## peek() works similarly to dequeue() but without popping.

# stack1 holds newly enqueued elements (most recent at top).
# stack2 holds elements in queue order with the queue front at its top.


class Queue2Stacks:

  # 1. Initialization
  def __init__(self):
    self.stack1 = StackLL()
    self.stack2 = StackLL()
    # Stack1 is for enqueueing
    # Stack2 is for dequeueing


  # 2. Check if queue is empty
  def isEmpty(self):
    return self.stack1.isEmpty() and self.stack2.isEmpty()


  # 3. Enqueue
  def enqueue(self, value):
    self.stack1.push(value)


  # 4. Dequeue
  def dequeue(self):
    if self.stack1.isEmpty() and self.stack2.isEmpty():
      return "Empty Queue"

    elif self.stack2.isEmpty():
      while not self.stack1.isEmpty():
        self.stack2.push(self.stack1.pop())


  # 5. Peek
  def peek(self):
    if self.isEmpty():
      return "Empty Queue"

    if self.stack2.isEmpty():
      while not self.stack1.isEmpty():
        self.stack2.push(self.stack1.pop())

    return self.stack2.peek()
```
