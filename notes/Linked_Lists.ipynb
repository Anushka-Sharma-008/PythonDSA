```python
# Creating a node

class Node:
  def __init__(self, value):
    self.data = value
    self.next = None
```

```python
# Creating individual nodes

a = Node(10)
b = Node(20)
c = Node(30)

print(a.data)
print(b.data)
print(c.data)

print(a.next)
print(b.next)
print(c.next)

print(a)
print(b)
print(c)
```

```python
# Connecting the individual nodes to form an LL

a.next = b
b.next = c

print(a.next)
print(b.next)
print(c.next)
```

```python
# Linked List Operations

class LinkedList:
  # 1. Initializing an empty LL (ie. LL with 0 nodes)
  # Empty LL -> Head = None
  def __init__(self):
    self.head = None
    self.n = 0
    # self.n -> Number of nodes in LL (Length)


  # 2. Length of LL
  def __len__(self):
    # Special method
    return self.n


  # 3. Insertion in LL
  # 3.1. Inserting head
  def insert_head(self, value):
    new_node = Node(value)
    # Creating new node

    new_node.next = self.head
    # Establishing connection

    self.head = new_node
    # Updating head

    self.n += 1
    # Updating length


  # 3.2. Inserting tail
  # Appending in LL
  def insert_tail(self, value):
    new_node = Node(value)
    # Creating new node

    # In case of empty LL
    if self.head == None:
      self.head = new_node
      self.n += 1
      # Updating length
      return

    # In case of non-empty LL
    current = self.head
    # Current m first node store kardia

    while current.next != None:
      # Meaning last node ka next is None, so vaha rukega

      current = current.next
    # Now you are at last node

    current.next = new_node
    # Establishing connection to the new node

    self.n += 1
    # Updating length


  # 3.3. Inserting after an item
  def insert_after(self, after, value):
    new_node = Node(value)
    # Creating new node

    current = self.head
    # Current m first node store kardia

    while current != None:
      if current.data == after:
        # Now you are at the node after which you want to insert. So break it
        break
      current = current.next

    # Case - 1: After vali value exists
    if current != None:
      # Bich m ghusa dia
      new_node.next = current.next
      current.next = new_node
      # After vale k next m jo agle vale ka address tha usko new m store krdia
      # Aur after vale k next m ab new_node ka address store krdia

      self.n += 1
      # Updating length

    # Case - 2: After vali value doesn't exist
    else:
      return "Item not found in LL"



  # 4. Traversal & Printing in LL
  def __str__(self):
    current = self.head
    # Current position (data and next)

    result = ''
    # Empty string which will store all the traversed items of the LL

    while current != None:
      # Here LL end tak jaega, aur jab khatam hojaega tab loop will end
      # Jaha pr rukna h vaha != None
      # If current se 1 pehle rukna h, then current.next != None
      # If current se 2 pehle rukna h, then current.next.next != None
      # If current se 3 pehle rukna h, then current.next.next.next != None
      # And so on

      result += str(current.data) + '->'
      current = current.next
    # Current position ka data print karke
    # Next k address ko current m store kardia

    return result[:-2]
    # Taaki last ka -> na print ho


  # 5. Deletion in LL
  # 5.1. Clearing an LL
  def clear(self):
    self.head = None
    # Bcz self.head = None mtlb empty LL

    self.n = 0
    # Updating length


  # 5.2. Delete from Head
  def delete_head(self):
    if self.head == None:
      return "LL is empty"
    # In case of empty LL

    self.head = self.head.next
    # Head k agle vale ko head m store krdia

    self.n -= 1
    # Updating length


  # 5.3. Delete from Tail (Pop)
  def pop(self):
    current = self.head

    # In case of empty LL
    if current == None:
      return "LL is empty"

    # In case of only 1 item
    if current.next == None:
      self.delete_head()
      return

    # All other cases
    while current.next.next != None:
      # Reaching for second last node
      current = current.next

    current.next = None
    # Second last node k next ko None bnado

    self.n -= 1
    # Updating length


  # 5.4. Delete by value (Remove)
  def remove(self, item):

    # In case of empty LL
    if self.head == None:
      return "LL is empty"

    # In case head ka hi data = item
    if self.head.data == item:
      self.delete_head()
      return

    current = self.head
    while current.next != None:
      # End tak traverse karo

      if current.next.data == item:
        # Current k next ka data item k equal ho agar then break the loop
        # If item nahi milega then current will eventually become the tail
        break
      current = current.next

    # Case - 1: If item is not found (Finally current.next == None)
    if current.next == None:
      return "Item not found in LL"

    # Case - 2: If item is found in LL
    else:
      current.next = current.next.next
      self.n -= 1
      # Updating length


  # 6. Searching in LL
  # 6.1. Searching based on item (Return index)
  def search_item(self, item):
    current = self.head
    index = 0

    while current != None:
      if current.data == item:
        return index
      current = current.next
      index += 1

    return "Item not found in LL"


  # 6.2. Searching based on index (Return item)
  def __getitem__(self, index):
    # Special method

    current = self.head
    position = 0

    while current != None:
      if position == index:
        return current.data
      current = current.next
      position += 1

    return "Index out of range"


  # 7. In-Place Reversal
  def reverse(self):
    previous = None
    # Starting point ka previous node

    current = self.head
    # Starting point

    while current != None:
      # End tak loop chalaenge

      next = current.next
      # Temporary variable aage k connection ko store krne k liye

      current.next = previous
      # Connection reverse kardia

      previous = current
      current = next

    self.head = previous


  # 8. Minimum
  def minimum(self):

    # In case of empty LL
    if self.head is None:
        return "LL is empty"

    current = self.head
    minimum = current

    while current != None:
      if current.data < minimum.data:
        minimum = current
      current = current.next

    return minimum


  # 9. Maximum
  def maximum(self):

    # In case of empty LL
    if self.head is None:
        return "LL is empty"
    current = self.head
    maximum = current

    while current != None:
      if current.data > maximum.data:
        maximum = current
      current = current.next

    return maximum


  # 10. Summation
  def summation(self):
    current = self.head
    total = 0

    while current != None:
      total += current.data
      current = current.next

    return total


  # 11. Sum of all nodes at odd places
  def sum_odd(self):
    current = self.head
    counter = 0
    total = 0

    while current != None:
      if counter % 2 != 0:
        total += current.data
      current = current.next
      counter += 1

    return total


  # 12. Sum of all nodes at even places
  def sum_even(self):
    current = self.head
    counter = 0
    total = 0

    while current != None:
      if counter % 2 == 0:
        total += current.data
      current = current.next
      counter += 1

    return total
```
